# Week 7: Line Plot of IoT Sensor Readings Over Time

In this notebook, we visualise IoT sensor readings over time using a line plot.  
We use the preprocessed dataset from Homework No. 3 and apply Seaborn and Matplotlib to show trends based on sensor type.


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned IoT data from the correct path
df = pd.read_csv("../data/cleaned_iot_data.csv")

# Preview first few rows
df.head()
```

```python
# Convert timestamp to datetime format
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Sort by timestamp to avoid zigzag plots
df = df.sort_values(by="timestamp")
```

```python
# Check unique data types (sensor types)
print("Sensor Types Found:", df["data_type"].unique())
```

```python
# Set visual style
sns.set(style="whitegrid")
```

```python
# Create the full line plot
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=df,
    x="timestamp",
    y="numeric_value",
    hue="data_type",       # differentiate sensor types
    marker="o"
)

# Formatting
plt.xticks(rotation=45)
plt.title("IoT Sensor Readings Over Time", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Sensor Value", fontsize=12)
plt.legend(title="Sensor Type")
plt.tight_layout()
plt.show()
```

```python
# 🕒 Zoom in to first 10 minutes to unclutter the plot
start_time = df["timestamp"].min()
end_time = start_time + pd.Timedelta(minutes=10)

zoomed_df = df[(df["timestamp"] >= start_time) & (df["timestamp"] <= end_time)]

# 🔍 Plot zoomed-in data
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=zoomed_df,
    x="timestamp",
    y="numeric_value",
    hue="data_type",
    marker="o"
)

plt.xticks(rotation=45)
plt.title("Zoomed View: IoT Sensor Readings (First 10 Minutes)", fontsize=14)
plt.xlabel("Timestamp", fontsize=12)
plt.ylabel("Sensor Value", fontsize=12)
plt.legend(title="Sensor Type")
plt.tight_layout()
plt.show()
```
